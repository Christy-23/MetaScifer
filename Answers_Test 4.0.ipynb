```python
q3=sales.quantile(0.75)
iqr=q3-q1
```

```python
#3- Find lower bound = Q1 - 1.5 × IQR
lower_bond = q1-(1.5*iqr)
```

```python
#4- Find upper bound = Q3 + 1.5 × IQR
upper_bond = q3+(1.5*iqr)
```
